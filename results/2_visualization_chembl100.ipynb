```python
data_info = {
    # default retrostar
    "0_chembl100_default_retrostar_0": result_path + "0_dpvn_ssm_chembl_100_subsample_all_routes_retrostar_0_policy_uspto_template_based_stock_emolecules_depth_30_cluster_top_100_routes_max_5_cluster.hdf",

    # default mcts
    "1_chembl100_default_mcts": result_path + "1_dpvn_ssm_chembl_100_subsample_all_routes_default_mcts_policy_uspto_template_based_stock_emolecules_depth_30_cluster_top_100_routes_max_5_cluster.hdf",

    # diversity forcing mcts
    "2_chembl100_diversity_forcing_mcts": result_path + "2_dpvn_ssm_chembl_100_subsample_all_routes_mcts_diversity_forcing_policy_uspto_template_based_stock_emolecules_depth_30_cluster_top_100_routes_max_5_cluster.hdf",    
    
    # expansion clustering mcts
    "3_chembl100_expansion_clustering_mcts": result_path + "3_dpvn_ssm_chembl_100_subsample_all_routes_expansion_clustering_mcts_first_occurence_policy_uspto_template_based_stock_emolecules_depth_30_cluster_top_100_routes_max_5_cluster.hdf",
    
    # expansion clustering retro*-0
    "4_chembl100_expansion_clustering_retrostar_0": result_path + "/5_dpvn_ssm_chembl_100_subsample_all_routes_retrostar_0_clustered_first_occurence_policy_uspto_template_based_stock_emolecules_depth_30_cluster_top_100_routes_max_5_cluster.hdf"    
    }
```

```python
# load data from hdf

import pandas as pd

def read_hdf(path):
    data = pd.read_hdf(path, key="table")
    # drop trees
    data = data.drop(columns=["trees"])
    profiling_df = pd.DataFrame(data["profiling"].tolist())
    data = pd.concat([data, profiling_df], axis=1)
    profiling_df = pd.DataFrame(data["policy_used_counts"].tolist())
    data = pd.concat([data, profiling_df], axis=1)
    
    return data

def calculate_solved_rate(df, maximum_expansion_calls = None):
    dataframe_length = len(df)
    if maximum_expansion_calls is not None:
        df = df[df["expansion_calls"] <= maximum_expansion_calls]
    solved_rate = df["is_solved"].sum() / dataframe_length
    return solved_rate

def calculate_solved_rate_based_on_iteration(df, maximum_iteration = None):
    dataframe_length = len(df)
    if maximum_iteration is not None:
        df = df[df["first_solution_iteration"] <= maximum_iteration]
    solved_rate = df["is_solved"].sum() / dataframe_length
    return solved_rate

selected_key = "3_chembl100_expansion_clustering_mcts"
selected_df = read_hdf(data_info[selected_key])
selected_df
```

```python
import numpy as np

# Initialize an empty list to store the results
results = []

# Iterate over each key in data_info
for key, path in data_info.items():
    if path != "":  # Check if the path is not empty
        print(f"Processing {key} in path {path}...")
        df = read_hdf(path)
        solved_rate = calculate_solved_rate(df)
        assert solved_rate == 1.0, f"Error: {key} is not fully solved"
        # calculate the solved rate between 0 and 50000 expansion calls
        
        expansion_iteration_range = []
        solved_rate_per_expansion_call = []
        
        for i in range(0, 25000, 100):
            expansion_iteration_range.append(i)
            solved_rate_per_expansion_call.append(calculate_solved_rate(df, i))

        iteration_range = []
        solved_rate_per_iteration = []
        
        for i in range(0, 600, 10):
            iteration_range.append(i)
            solved_rate_per_iteration.append(calculate_solved_rate_based_on_iteration(df, i))

        average_solved_iteration = df["first_solution_iteration"].mean()        
        average_expansion_calls = df["expansion_calls"].mean()
        average_search_time = df["search_time"].mean() / 60 
        search_time = df["search_time"].to_list()
        mean_number_of_steps = df["number_of_steps"].mean()

        # distance matrix
        distance_matrices = [np.array(matrix) for matrix in df["distance_matrix"]]
        distance_matrix_means = [np.mean(matrix) for matrix in distance_matrices]
        distance_matrix_means_mean = np.mean(distance_matrix_means)

        if "clustered_tree_width" in df.columns:
            clustered_tree_width = df["clustered_tree_width"].apply(
                lambda x: sum(x) / len(x) if len(x) > 0 else 0
                ).mean()
        else:
            clustered_tree_width = 0

        if "tree_width" in df.columns:
            normal_tree_width = df["tree_width"].apply(
                lambda x: sum(x) / len(x) if len(x) > 0 else 0
                ).mean()
        else:
            normal_tree_width = 0

        mean_number_of_solved_routes = df["number_of_solved_routes"].mean()
        mean_number_of_route_clusters = df["cluster_labels"].apply(lambda x: len(set(x))).mean()

        # full info
        number_of_solved_routes = df["number_of_solved_routes"].to_list()
        number_of_route_clusters = df["cluster_labels"].apply(lambda x: len(set(x))).to_list()
        number_of_steps = df["number_of_steps"].to_list()

        # route statistics
        current_statistics = dfs_dict[key]

        # get the minimum reaction count
        minimum_reaction_count_per_target = current_statistics.groupby('target_smiles')['reaction_count'].min().reset_index()
        minimum_reaction_count_per_target = minimum_reaction_count_per_target["reaction_count"].values.tolist()

        assert min(minimum_reaction_count_per_target) > 0, f"Error: {key} has minimum reaction count equal than 0"

        all_reaction_counts = current_statistics['reaction_count'].values.tolist()

        assert min(all_reaction_counts) > 0, f"Error: {key} has minimum reaction count equal than 0"

        stats_solved_routes = current_statistics.groupby('target_smiles')['solved'].sum().reset_index()["solved"].values.tolist()
        assert min(stats_solved_routes) > 0, f"Error: {key} has minimum stats_solved_routes to 0"

        result_dict = {
            "key": key,
            "solved_rate": solved_rate,
            "expansion_iteration_range": expansion_iteration_range,
            "solved_rate_per_expansion_call": solved_rate_per_expansion_call,
            "iteration_range": iteration_range,
            "solved_rate_per_iteration": solved_rate_per_iteration,
            "average_solved_iteration": average_solved_iteration,
            "average_search_time": average_search_time,
            "search_time": search_time,
            "average_expansion_calls": average_expansion_calls,
            "mean_number_of_steps": mean_number_of_steps,
            # number of solved routes
            "mean_number_of_solved_routes": mean_number_of_solved_routes,
            "mean_number_of_route_clusters": mean_number_of_route_clusters,
            # tree width
            "average_clustered_tree_width": clustered_tree_width,
            "average_normal_tree_width": normal_tree_width,
            # full stats
            "number_of_solved_routes": number_of_solved_routes,
            "number_of_route_clusters": number_of_route_clusters,
            "number_of_steps": number_of_steps,
            # distane matrix means
            "top_100_routes_distance": distance_matrix_means,
            "mean_top100_routes_distance": distance_matrix_means_mean,
            # route statistcs data
            "route_stats_minimum_reaction_count_per_target": minimum_reaction_count_per_target,
            "route_stats_all_reaction_counts": all_reaction_counts,
            "route_stats_solved_routes": stats_solved_routes
        }

        #average_policy_used_counts = df["policy_used_counts_retrostar_mlp"].mean()
        results.append(result_dict)

# Create a new dataframe from the results
solved_rate_df = pd.DataFrame(results)
solved_rate_df
```

```python
# split keyi nto dataset and algorithm
solved_rate_df["dataset"] = solved_rate_df["key"].apply(lambda x: x.split("_")[1])

# algoirthm options are: "retrostar", "DEFAULT", "SUPER". so a string comparison
def find_algorithm(key):
    if "0_chembl100_default_retrostar_0" in key:
        return "Retro*-0"
    elif "1_chembl100_default_mcts" in key:
        return "MCTS"
    elif "2_chembl100_diversity_forcing_mcts" in key:
        return "Distance-MCTS"
    elif "3_chembl100_expansion_clustering_mcts" in key:
        # expansion_clustering_mcts
        return "Cluster-MCTS"
    elif "4_chembl100_expansion_clustering_retrostar_0" in key:
        # expansion_clustering_retrostar
        return "Cluster-Retro*-0"

solved_rate_df["algorithm"] = solved_rate_df["key"].apply(find_algorithm)
solved_rate_df
```

```python
number_of_solved_routes = explode_subset.explode('number_of_solved_routes').loc[:, ['algorithm', 'number_of_solved_routes']].reset_index(drop=True)
# number_of_steps
number_of_steps = explode_subset.explode('number_of_steps').loc[:, ['algorithm', 'number_of_steps']].reset_index(drop=True)
# top_100_routes_distance
top_100_routes_distance = explode_subset.explode('top_100_routes_distance').loc[:, ['algorithm', 'top_100_routes_distance']].reset_index(drop=True)
# route_stats_minimum_reaction_count_per_target
route_stats_minimum_reaction_count_per_target = explode_subset.explode('route_stats_minimum_reaction_count_per_target').loc[:, ['algorithm', 'route_stats_minimum_reaction_count_per_target']].reset_index(drop=True)
#route_stats_all_reaction_counts
route_stats_all_reaction_counts = explode_subset.explode('route_stats_all_reaction_counts').loc[:, ['algorithm', 'route_stats_all_reaction_counts']].reset_index(drop=True)
#route_stats_solved_routes
route_stats_solved_routes = explode_subset.explode('route_stats_solved_routes').loc[:, ['algorithm', 'route_stats_solved_routes']].reset_index(drop=True)
route_stats_solved_routes
```

```python
# order: stat solved routes, shortest route found, top 100 routes distance

import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set(style="white")

algorithm_order = [
    'MCTS',
    'Distance-MCTS',
    'Cluster-MCTS',
    'Retro*-0',
    'Cluster-Retro*-0'
]

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=False)
# Define a color palette
palette = sns.color_palette("bright", len(algorithm_order))

# Box plot for solved routes
sns.boxplot(data=route_stats_solved_routes, x="algorithm", y="route_stats_solved_routes", ax=axes[0], palette=palette, order=algorithm_order)
axes[0].set_ylabel("Number of Solved Routes")
axes[0].tick_params(axis='x', rotation=45)

# Box plot for minimum reaction count per target
sns.boxplot(data=route_stats_minimum_reaction_count_per_target, x="algorithm", y="route_stats_minimum_reaction_count_per_target", ax=axes[1], palette=palette, order=algorithm_order)
axes[1].set_ylabel("Number of Reactions in Shortest Route")
axes[1].set_yticks(range(0, 21, 2))
axes[1].tick_params(axis='x', rotation=45)

# Box plot for top 100 routes distance
sns.boxplot(data=top_100_routes_distance, x="algorithm", y="top_100_routes_distance", ax=axes[2], palette=palette, order=algorithm_order)
axes[2].set_ylabel("Avg. Distance between 100 Shortest Routes")
axes[2].tick_params(axis='x', rotation=45)

# remove x axis description and call it "Search Algorithm" in the middle plot
axes[0].set_xlabel("")
axes[1].set_xlabel("Search Algorithm")
axes[2].set_xlabel("")

# Despine and show
sns.despine()
plt.tight_layout()
plt.savefig(f"{save_path}/3_100_chembl_molecules_emolecules_stock.pdf", bbox_inches='tight')
plt.show()
```
