```python
from rdkit import Chem

# Function to load a reaction from a SMILES string and return the reaction object
def load_reaction(reaction_smiles):
    reaction = Chem.rdChemReactions.ReactionFromSmarts(reaction_smiles)
    return reaction
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
from rdkit.Chem.Draw import rdMolDraw2D, rdDepictor
from IPython.display import SVG, display
import sys

# Set CoordGen as preferred 2D coordinate generator for better layout
rdDepictor.SetPreferCoordGen(True)

def render_reaction_acs_style(reaction_smiles, width=600, height=300):
    """
    Creates an ACS-style reaction diagram with proper publication drawing settings,
    displaying larger molecules first with color atoms, while handling non-kekulizable structures
    """
    # Parse the reaction
    rxn = Chem.rdChemReactions.ReactionFromSmarts(reaction_smiles)
    
    # Create a new reaction with optimized coordinates
    new_rxn = AllChem.ChemicalReaction()
    
    # Process products - sort by size (largest first)
    products = []
    for mol in rxn.GetProducts():
        # Try different approaches to get a valid molecule
        try:
            # Convert to SMILES then back with specific sanitization flags that skip kekulization
            smiles = Chem.MolToSmiles(mol)
            mol = Chem.MolFromSmiles(smiles, sanitize=False)
            
            if mol is None:
                continue
                
            # Apply custom sanitization that skips the kekulization step
            Chem.SanitizeMol(mol, 
                sanitizeOps=Chem.SanitizeFlags.SANITIZE_ALL^Chem.SanitizeFlags.SANITIZE_KEKULIZE^Chem.SanitizeFlags.SANITIZE_PROPERTIES^Chem.SANITIZE_SETAROMATICITY)
                
        except Exception as e:
            print(f"Error processing product: {e}")
            continue
            
        # Add to products list with size info
        products.append((mol, mol.GetNumAtoms()))
    
    # Sort products by size (largest first)
    products.sort(key=lambda x: x[1], reverse=True)
    
    # Add sorted products to reaction
    for mol, _ in products:
        # Try to compute coordinates safely
        try:
            AllChem.Compute2DCoords(mol)
        except Exception as e:
            # If that fails, try with CoordGen directly
            try:
                rdDepictor.Compute2DCoords(mol)
            except Exception as e2:
                print(f"Error computing coordinates: {e2}")
            
        new_rxn.AddProductTemplate(mol)
    
    # Process reactants - sort by size (largest first) 
    reactants = []
    for mol in rxn.GetReactants():
        try:
            # Convert to SMILES then back with specific sanitization flags that skip kekulization
            smiles = Chem.MolToSmiles(mol)
            mol = Chem.MolFromSmiles(smiles, sanitize=False)
            
            if mol is None:
                continue
                
            # Apply custom sanitization that skips the kekulization step
            Chem.SanitizeMol(mol, 
                sanitizeOps=Chem.SanitizeFlags.SANITIZE_ALL^Chem.SanitizeFlags.SANITIZE_KEKULIZE^Chem.SanitizeFlags.SANITIZE_PROPERTIES^Chem.SANITIZE_SETAROMATICITY)
                
        except Exception as e:
            print(f"Error processing reactant: {e}")
            continue
            
        # Add to reactants list with size info
        reactants.append((mol, mol.GetNumAtoms()))
    
    # Sort reactants by size (largest first)
    reactants.sort(key=lambda x: x[1], reverse=True)
    
    # Add sorted reactants to reaction
    for mol, _ in reactants:
        # Try to compute coordinates safely
        try:
            AllChem.Compute2DCoords(mol)
        except Exception as e:
            # If that fails, try with CoordGen directly
            try:
                rdDepictor.Compute2DCoords(mol)
            except Exception as e2:
                print(f"Error computing coordinates: {e2}")
            
        new_rxn.AddReactantTemplate(mol)
    
    # Create drawer
    drawer = rdMolDraw2D.MolDraw2DSVG(width, height)
    opts = drawer.drawOptions()
    
    # Apply ACS 1996 style
    if reactants:
        # Get mean bond length from first (largest) reactant to set proper scaling
        largest_reactant = reactants[0][0]
        Draw.SetACS1996Mode(opts, Draw.MeanBondLength(largest_reactant))

    
    # Additional drawing options
    opts.bondLineWidth = 1
    
    # Do not attempt to sanitize or kekulize molecules during drawing
    opts.prepareMolsBeforeDrawing = True
    
    # Draw the reaction
    drawer.DrawReaction(new_rxn)
    drawer.FinishDrawing()
    svg = drawer.GetDrawingText()
    
    return SVG(svg)

# Function to display a reaction
def display_reaction_acs(reaction_obj_or_smiles):
    if isinstance(reaction_obj_or_smiles, str):
        # It's a SMILES string
        return render_reaction_acs_style(reaction_obj_or_smiles)
    else:
        # It's a reaction object
        rxn_smiles = Chem.rdChemReactions.ReactionToSmarts(reaction_obj_or_smiles)
        return render_reaction_acs_style(rxn_smiles)

# Test with the first reaction
test_reaction = reactions[0]
display(display_reaction_acs(test_reaction))
```

```python
# Visualize reactions by cluster label
import os
from rdkit import Chem

def save_svg_to_file(svg_text, filepath):
    """
    Save SVG content to a file
    
    Parameters:
    -----------
    svg_text : str
        The SVG content as a string
    filepath : str
        The path where to save the SVG file
    """
    with open(filepath, 'w') as f:
        f.write(svg_text)
    print(f"SVG saved to {filepath}")

# Make sure save_path is defined before this code
# Create a directory for each cluster if needed
for cluster in sorted(set(labels)):  # Sort unique cluster labels
    print(f"## Cluster {cluster}")
    print("-" * 50)
    
    # Create cluster-specific directory
    cluster_save_path = f"{save_path}/cluster_{cluster}"
    os.makedirs(cluster_save_path, exist_ok=True)
    
    # Find reactions in this cluster
    for i in range(len(reactions)):
        if labels[i] == cluster:
            current_reaction = reactions[i]
            reaction = load_reaction(current_reaction)
            print(f"Reaction: {i}")
            print(f"Importance: {importance[i]}")
            print(f"Label: {labels[i]}")
            print(f"Reaction: {current_reaction}")
            
            # Display the reaction
            display(display_reaction_acs(reaction))
            
            # Save the reaction as a picture with the reaction name
            if save_images:
                # Format the filename with reaction name and metadata
                reaction_name = current_reaction.replace(">>", "_")
                reaction_name = f"{reaction_name}_importance_{importance[i]:.2f}_cluster_{cluster}"
                
                # Save the reaction image
                #reaction_picture = Chem.Draw.ReactionToImage(display_reaction_acs(reaction))
                reaction_svg = display_reaction_acs(reaction)
                svg_text = reaction_svg.data
                save_path_svg = f"{cluster_save_path}/{reaction_name}.svg"
                save_svg_to_file(svg_text, save_path_svg)
                #reaction_picture.save(f"{cluster_save_path}/{reaction_name}.png")
                print(f"Saved image to {cluster_save_path}/{reaction_name}.svg")
            
            print("-" * 30)
    
    print("\n")
```
